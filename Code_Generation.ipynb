````python
# Setup koneksi ke Groq API menggunakan API key

from groq import Groq
from getpass import getpass

api_key = getpass("Masukkan Groq API Key: ")
client = Groq(api_key=api_key)

print("Groq client berhasil dibuat!")
````

````python
# Fungsi utama: terima deskripsi bahasa Indonesia, kembalikan kode Python

def generate_code(deskripsi: str) -> str:
    prompt = f"""Kamu adalah asisten code generation.
Tugasmu hanya menulis kode Python berdasarkan deskripsi berikut.

Aturan:
- Hanya tulis kode Python, tanpa penjelasan
- Tambahkan komentar singkat di dalam kode jika perlu
- Jangan tambahkan teks sebelum atau sesudah kode

Deskripsi: {deskripsi}
"""
    response = client.chat.completions.create(
        model="llama-3.3-70b-versatile",
        messages=[{"role": "user", "content": prompt}],
        temperature=0.2,
    )
    return response.choices[0].message.content.strip()
````

````python
# Uji coba fungsi dengan beberapa contoh deskripsi

contoh = [
    "buat fungsi cek bilangan prima",
    "buat fungsi hitung faktorial dengan rekursi",
    "buat fungsi balik urutan string",
]

for deskripsi in contoh:
    print(f"INPUT : {deskripsi}")
    print(f"OUTPUT:\n{generate_code(deskripsi)}")
    print("-" * 50)
````

````python
# Bersihkan markdown fence dari output agar kode siap digunakan dan buat backrgound hitam agar terlihat seperti source code asli dan bukan text yang menyamar

import re
from IPython.display import display, HTML

def clean_code(raw: str) -> str:
    return re.sub(r"```(?:python)?|```", "", raw).strip()

def generate_clean_code(deskripsi: str) -> str:
    raw = generate_code(deskripsi)
    return clean_code(raw)

# Test ulang dengan output yang sudah bersih
for deskripsi in contoh:
    print(f"INPUT : {deskripsi}")
    display(HTML(f"""
<pre style="background-color:#1e1e1e; color:#d4d4d4; padding:16px; border-radius:8px; font-family:monospace; font-size:14px; overflow-x:auto;">
{generate_clean_code(deskripsi)}
</pre>
"""))
    print("-" * 50)
````

````python
# Menu interaktif untuk memilih use case code generation

from IPython.display import display, HTML

def generate_by_usecase(pilihan, user_input):
    prompts = {
        "1": f"Kamu adalah asisten code generation. Tulis kode Python berdasarkan input berikut. Input bisa berupa kalimat biasa atau komentar kode Python (diawali #). Hanya tulis kode, tanpa penjelasan.\n\nInput: {user_input}",
        "2": f"Kamu adalah asisten SQL. Ubah kalimat berikut menjadi query SQL. Hanya tulis query SQL, tanpa penjelasan.\n\nKalimat: {user_input}",
        "3": f"Kamu adalah asisten refactoring. Perbaiki dan rapikan kode berikut tanpa mengubah fungsinya. Hanya tulis kode hasil refactor, tanpa penjelasan.\n\nKode: {user_input}",
        "4": f"Kamu adalah asisten testing. Buatkan unit test Python menggunakan unittest untuk fungsi berikut. Hanya tulis kode unit test, tanpa penjelasan.\n\nFungsi: {user_input}",
        "5": f"Kamu adalah asisten dokumentasi. Tambahkan docstring dan komentar yang jelas pada kode berikut. Hanya tulis kode dengan dokumentasi, tanpa penjelasan tambahan.\n\nKode: {user_input}",
        "6": "Kamu adalah asisten code generation. Lanjutkan kode Python berdasarkan komentar berikut. Hanya tulis kode, tanpa penjelasan.\n\n# Fungsi untuk menghitung Fibonacci ke-n dengan memoization",
    }

    response = client.chat.completions.create(
        model="llama-3.3-70b-versatile",
        messages=[{"role": "user", "content": prompts[pilihan]}],
        temperature=0.2,
    )
    raw = response.choices[0].message.content.strip()
    return clean_code(raw)

print("Pilih use case:")
print("1. Generate fungsi Python dari deskripsi")
print("2. Natural Language → SQL")
print("3. Refactoring kode")
print("4. Generate unit test")
print("5. Generate dokumentasi kode")
print("6. Autocomplete dari komentar (demo)")

pilihan = input("\nMasukkan nomor pilihan (1-6): ")

if pilihan == "6":
    user_input = ""
else:
    user_input = input("Masukkan input kamu: ")

hasil = generate_by_usecase(pilihan, user_input)

display(HTML(f"""
<pre style="background-color:#1e1e1e; color:#d4d4d4; padding:16px; border-radius:8px; font-family:monospace; font-size:14px; overflow-x:auto;">
{hasil}
</pre>
"""))
````

````python
# Demo autocomplete: AI melanjutkan kode dari komentar di bawah ini
# Fungsi untuk menghitung Fibonacci ke-n dengan memoization

hasil_demo = generate_by_usecase("6", "")

display(HTML(f"""
<pre style="background-color:#1e1e1e; color:#d4d4d4; padding:16px; border-radius:8px; font-family:monospace; font-size:14px; overflow-x:auto;">
{hasil_demo}
</pre>
"""))
````
